# 2D Scatter plot

```python
#2D SCATTER PLOT
haberman.plot(kind='scatter',x='age',y="nodes")
plt.grid()
plt.title('scatter plot')
plt.show()
```

```python
# 2-D Scatter plot with color-coding for each type/class.
# Here 'sns' corresponds to seaborn. 
sns.set_style("whitegrid");
sns.FacetGrid(haberman,hue="status",size=6).map(plt.scatter, "age", "nodes").add_legend()
sns.FacetGrid(haberman,hue="status",size=6).map(plt.scatter, "status", "nodes").add_legend()
sns.FacetGrid(haberman,hue="status",size=6).map(plt.scatter, "age", "status").add_legend()
plt.show()

# Notice that the blue and orange points can be easily seperated  
# we can draw multiple 2-D scatter plots for each combination of features
# we are importing warning here to clear warning
import warnings
warnings.filterwarnings("ignore")
# here we are importing warning 
```

# Pair plots

```python
#pairplot
# pairwise scatter plot: Pair-Plot :3*3 plot
# Dis-advantages: 
##Can be used when number of features are high.
##Cannot visualize higher dimensional patterns in 3-D and 4-D. 
#Only possible to view 2D patterns.
plt.close();
sns.set_style("whitegrid")
sns.pairplot(haberman,hue="status",size=4,vars=['age','year','nodes'])
plt.title('pair_plots')
plt.show();
# here we are importing warning 
```

# PDF CDF

```python
sns.FacetGrid(haberman,hue='status',size=5).map(sns.distplot,"nodes").add_legend();
plt.title('PDF of nodes')
sns.FacetGrid(haberman,hue='status',size=5).map(sns.distplot,"age").add_legend();
plt.title('PDF of age')
sns.FacetGrid(haberman,hue='status',size=5).map(sns.distplot,"year").add_legend();
plt.title('PDF of year')
plt.show();
```

#Observation
#It has observed that people survive long if they have less nodes detected viceversa
#From  above PDF we can say the patients survival status
#To say what percentage of people actually small survive or large survive , we have another distribution called as CDF. 

```python
# cdf pdf
#plot cdf of age and year

#age
counts, bin_edges = np.histogram(haberman['status'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf);
print(bin_edges);
plt.plot(bin_edges[1:],pdf);
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf)
plt.title('age')
```

# Mean, Variance, SD

```python
#mean, variance, SD

print("Mean:")
print(np.mean(haberman["age"]))
print(np.mean(haberman["year"]))
print(np.mean(haberman["nodes"]))
print(np.mean(haberman["status"]))

#SD
print("\n standard_deviation:")
print(np.std(haberman["age"]))
print(np.std(haberman["year"]))
print(np.std(haberman["nodes"]))
print(np.std(haberman["status"]))
```

# Median, Percentile, Quantile, MAD

```python
#median, percentile, quantile, MAD
print("\n median")
print(np.median(haberman["age"]))
print(np.median(haberman["year"]))
print(np.median(haberman["nodes"]))
print(np.median(haberman["status"]))

#quantile
print("\n quantile:")
print(np.percentile(haberman["age"],np.arange(0,100,25)))
print(np.percentile(haberman["year"],np.arange(0,100,25)))
print(np.percentile(haberman["nodes"],np.arange(0,100,25)))
print(np.percentile(haberman["status"],np.arange(0,100,25)))

#90th percentile
print("\n 90th percentile")
print(np.percentile(haberman["age"],90))
print(np.percentile(haberman["year"],90))
print(np.percentile(haberman["nodes"],90))
print(np.percentile(haberman["status"],90))

#MAD
from statsmodels import robust
print("\n median absolute deviation")
print(robust.mad(haberman["age"]))
print(robust.mad(haberman["year"]))
print(robust.mad(haberman["nodes"]))
print(robust.mad(haberman["status"]))
```

# Box plot

# Contour plot

```python
#contour plot
#2D density plot
#contour plot are like density plot 

sns.jointplot(x="age",y="status",data=haberman,kind="kde")
plt.title('contour plot')
plt.show()
```

```python
sns.jointplot(x="age",y="nodes",data=haberman,kind="kde")
plt.title('contour plot')
plt.show()
```

#observation:
#Above is the 2D density plot for long survival using feature age and nodes,
#It is obswerved the density of point for long survival is more from age range 47-60 and nodes from 0-3.
